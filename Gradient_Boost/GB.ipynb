```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
X = np.random.rand(100, 1) - 0.5  
y = 3 * X[:, 0]**2+0.05 * np.random.randn(100)
```

```python
df=pd.DataFrame()
```

```python
df['X']=X.reshape(100)
df['y']=y
```

```python
df
```

```python
plt.scatter(df['X'],df['y'])
plt.title('X vs Y')
```

```python
df['prediction1']=df['y'].mean()
df
```

```python
df['residual1']=df['y']-df['prediction1']
df
```

```python
plt.scatter(df['X'],df['y'])
plt.plot(df['X'],df['prediction1'],color='purple')
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree1=DecisionTreeRegressor(max_leaf_nodes=8)
```

```python
tree1.fit(df['X'].values.reshape(100,1),df['residual1'].values)
```

```python
from sklearn.tree import plot_tree
plot_tree(tree1)
plt.show()
```

```python
X_test = np.linspace(-0.5,0.5,500)
```

```python
Y_pred = 0.265458+tree1.predict(X_test.reshape(500,1))
```

```python
plt.figure(figsize=(14,4))
plt.subplot(1,2,1)
plt.plot(X_test,Y_pred,linewidth=2,color='orange')
plt.scatter(df['X'],df['y'])
```

```python
df['prediction2'] = 0.265458 + tree1.predict(df['X'].values.reshape(100,1))
df
```

```python
df['residual2']=df['y']-df['prediction2']
df
```

```python
tree2=DecisionTreeRegressor(max_leaf_nodes=8)
```

```python
tree2.fit(df['X'].values.reshape(100,1),df['residual2'].values)
```

```python
def gradient_boost(X,y,number,lr,count=1,regs=[],foo=None):

  if number == 0:
    return 
  else:
    # do gradient boosting

    if count > 1:
      y = y - regs[-1].predict(X)
    else:
      foo = y

    tree_reg = DecisionTreeRegressor(max_depth=5, random_state=42)
    tree_reg.fit(X, y)

    regs.append(tree_reg)

    x1 = np.linspace(-0.5, 0.5, 500)
    y_pred = sum(lr * regressor.predict(x1.reshape(-1, 1)) for regressor in regs)

    print(number)
    plt.figure()
    plt.plot(x1, y_pred, linewidth=2)
    plt.plot(X[:, 0], foo,"r.")
    plt.show()

    gradient_boost(X,y,number-1,lr,count+1,regs,foo=foo)
```

```python
np.random.seed(42)
X = np.random.rand(100, 1) - 0.5
y = 3*X[:, 0]**2 + 0.05 * np.random.randn(100)
gradient_boost(X,y,5,lr=1)
```
